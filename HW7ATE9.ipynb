The Simple Linear Regression (SLR) model is a method to explore and quantify the relationship between two continuous variables, usually referred to as the predictor and the outcome. SLR aims to fit a straight line to observed data points in such a way that we can predict or understand changes in the outcome variable based on changes in the predictor variable.

Components of the Simple Linear Regression Model
Predictor Variable (x): This is the independent variable. We think this variable influences or explains the outcome. For example, if we're studying the impact of study hours on test scores, "study hours" would be our predictor.

Outcome Variable (y): This is the dependent variable, or the one we're trying to predict or explain. In the same example, "test scores" would be our outcome variable.

Slope Coefficient (β1): The slope indicates how much the outcome variable changes with a one-unit increase in the predictor variable. If the slope is positive, the relationship is direct: as x increases, y also increases. If it’s negative, as x increases, y decreases. This coefficient tells us the strength and direction of the relationship.

Intercept Coefficient (β0): The intercept is the expected value of y when x equals zero. It’s where the regression line crosses the y-axis. For example, if "study hours" were zero, the intercept would represent the expected baseline test score.

Error Term (ε): Real-world data is rarely perfectly linear, so there is an error term to account for the random variation in the outcome that the predictor cannot explain. This term represents the difference between the actual observed values and the predicted values along the regression line.

How They Combine to Form the Model
In the Simple Linear Regression model, these elements come together in the equation: 
y = β0 + β1x + ε

Fited Line (Blue, Solid): This line is got by applying linear regression to the simulated dataset using ols. Its slope and intercept are estimated based on the data, which is influenced by random sampling variation. Therefore, the fitted line may deviate slightly from the theoretical line each time the dataset is regenerated due to inherent randomness in the error term.

Key Difference:
The theoretical line shows the ideal or expected relationship in a population, while the fitted line demonstrates the relationship as etsimated from a sample, which can vary slightly due to random sampling variation. This difference highlights how sample data may not perfectly align with theoretical models due to random variability, even when the model assumptions hold true.

In pythons we can access these values using these:
fitted_model.params['Intercept'] corresponds to 𝛽0,

Thus, the fitted values 𝑦i for each observation 𝑥𝑖 can be computed by substituting the estimated parameters:

fitted_model.summary().tables[1] displays the estimated parameters (coefficients) under the "Coef" column, which are used to compute fitted values yi for each xi in the data.

fitted_model.fittedvalues leverages these coefficients to produce the fitted values, generating in-sample predictions specific to the data used for fitting.

The line chosen for the fitted model using the "ordinary least squares" (OLS) method is the line that minimizes the total squared residuals — the differences between observed values (Yi) and the values predicted by the model (yi). Here, residuals (𝜖^i = Yi - yi) are distinct from the theoretical model's error terms (𝜖𝑖), as residuals specifically represent deviations in the fitted model based on obsevred data.

In the visualization, vertical dashed lines highlight each residual. The OLS line is chosen so that the sum of the squares of these residuals is as small as possible. Squaring the residuals is necessary because it:

Emphasizes larger errors (larger deviations contribute more to the total error),
Prevents positive and negative residuals from canceling each other out, ensuring an accurate minimization of overall prediction error. The result is a fitted line that optimally predicts Y based on x with the smallest possible deviations (residuals) from observed values.

```python
import numpy as np
import statsmodels.formula.api as smf
import plotly.express as px

# Filter data for long wait times
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Initialize array to store bootstrap slope coefficients
num_bootstrap_samples = 1000
bootstrapped_slope_coefficients = []

# Perform bootstrap sampling
for _ in range(num_bootstrap_samples):
    bootstrap_sample = old_faithful[long_wait].sample(n=long_wait.sum(), replace=True)
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrapped_slope_coefficients.append(model.params['waiting'])

# Convert to a NumPy array for easier manipulation
bootstrapped_slope_coefficients = np.array(bootstrapped_slope_coefficients)

# Plot the bootstrap distribution of the slope coefficients
px.histogram(bootstrapped_slope_coefficients, title='Bootstrap Distribution of Slope Coefficients').show()


##now we cn set parameters and generate data
from scipy import stats

# Create simulated dataset under the null hypothesis
old_faithful_simulation = old_faithful[long_wait].copy()
old_faithful_simulation['duration'] = 1.65 + 0 * old_faithful_simulation.waiting + stats.norm.rvs(loc=0, scale=0.37, size=long_wait.sum())

# Initialize array to store simulated slope coefficients
simulated_slope_coefficients = []

# Perform simulations
for _ in range(num_bootstrap_samples):
    simulated_sample = old_faithful_simulation.sample(n=long_wait.sum(), replace=True)
    model = smf.ols('duration ~ waiting', data=simulated_sample).fit()
    simulated_slope_coefficients.append(model.params['waiting'])

# Convert to a NumPy array for easier manipulation
simulated_slope_coefficients = np.array(simulated_slope_coefficients)

# Plot the simulated distribution of the slope coefficients
px.histogram(simulated_slope_coefficients, title='Simulated Distribution of Slope Coefficients Under Null Hypothesis').show()

# Calculate 95% bootstrapped confidence interval
bootstrap_conf_interval = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])

# Calculate the mean of simulated slope coefficients for comparison
simulated_mean_slope = np.mean(simulated_slope_coefficients)

# Check if zero is within the bootstrap confidence interval
zero_in_confidence_interval = 0 >= bootstrap_conf_interval[0] and 0 <= bootstrap_conf_interval[1]

# Calculate the p-value for the simulated distribution
simulated_p_value = (np.abs(simulated_slope_coefficients) >= np.abs(simulated_mean_slope)).mean()

print("95% Bootstrapped Confidence Interval for Slope:", bootstrap_conf_interval)
print("Is Zero within the Confidence Interval?:", zero_in_confidence_interval)
print("Simulated Mean Slope under Null Hypothesis:", simulated_mean_slope)
print("Simulated P-value:", simulated_p_value)
```

Part (a): Explain How the Sampling Approaches Work for the Two Simulations
Permutation Test:

The permutation test works by randomly shuffling the labels "short" and "long" associated with each wait time duration, effectively breaking any original relationship between wait time type and duration.
After shuffling, the mean duration difference between the shuffled "short" and "long" groups is calculated. This process is repeated multiple times to create a distribution of mean differences that would be expected under the null hypothesis (i.e., no real difference between the groups).
The observed mean difference from the original data is then compared to this distribution to see if it is unusual (i.e., if it falls in the tail ends of the permutation distribution), providing a p-value as evidence against the null hypothesis.
Bootstrap Confidence Interval:

The bootstrap approach resamples the original data within each group ("short" and "long") with replacement to generate multiple samples for each group.
For each bootstrap sample, the mean duration for the "short" group and the mean duration for the "long" group are calculated, and their difference is recorded.
This process is repeated many times, creating a distribution of mean differences between the groups. A 95% confidence interval is then created by taking the 2.5th and 97.5th percentiles of this bootstrapped distribution of mean differences.
This confidence interval provides an estimate of the range within which we expect the true mean difference to lie, based on the variability in the original data.
Part (b): Comparison with Indicator Variable Based Model (Question 11)
Similarity:

Both the permutation test and the indicator variable model from Question 11 aim to assess whether there is a significant difference between the "short" and "long" wait groups. The permutation test does this through random shuffling, while the indicator variable model compares the mean durations by including an indicator for the group in a linear model.
Both methods are also non-parametric in their reliance on resampling or shuffling rather than assumptions of normality.
Differences:

The permutation and bootstrap tests involve resampling or shuffling to build a distribution of differences in means, while the indicator variable model directly estimates the difference between groups through regression, assuming the error terms are normally distributed.
The indicator variable model provides a single estimate of the difference with an associated confidence interval derived from the regression framework, while the permutation and bootstrap tests rely on empirical distributions generated from the data itself.
The permutation test and bootstrap confidence interval offer a more flexible approach that doesn't assume a specific distribution for the error terms, making them useful when we’re unsure about normality assumptions. The indicator model, however, uses a straightforward linear regression and assumes normality, making it efficient but potentially less robust if the normality assumption does not hold.
