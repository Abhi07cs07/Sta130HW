```python
#Answer 4: 

Let's break down how to produce a 95% bootstrap confidence interval for the population mean and see how we can adjust it for other statistics, like the population median.

Here’s Python code to generate a 95% bootstrap confidence interval for the mean, with comments explaining how to modify it for other statistics (like the median).

python
Copy code
import numpy as np

# Sample data (replace with your own sample)
sample = np.array([10, 20, 30, 40, 50])

# Number of bootstrap samples
n_bootstraps = 10000

# Empty array to store bootstrap sample means
bootstrap_means = []

# Resample with replacement and calculate the mean for each bootstrap sample
for _ in range(n_bootstraps):
    bootstrap_sample = np.random.choice(sample, size=len(sample), replace=True)  # Bootstrap sample
    bootstrap_mean = np.mean(bootstrap_sample)  # Calculate mean of the bootstrap sample
    bootstrap_means.append(bootstrap_mean)

# Calculate the 95% confidence interval
lower_bound = np.percentile(bootstrap_means, 2.5)  # 2.5th percentile
upper_bound = np.percentile(bootstrap_means, 97.5)  # 97.5th percentile

print(f"95% Confidence Interval for the Mean: [{lower_bound}, {upper_bound}]")

# To calculate a 95% confidence interval for the median:
# 1. Change the calculation of the bootstrap statistic to the median
# 2. Use np.median instead of np.mean

bootstrap_medians = []

for _ in range(n_bootstraps):
    bootstrap_sample = np.random.choice(sample, size=len(sample), replace=True)  # Bootstrap sample
    bootstrap_median = np.median(bootstrap_sample)  # Calculate median of the bootstrap sample
    bootstrap_medians.append(bootstrap_median)

# Calculate the 95% confidence interval for the median
lower_bound_median = np.percentile(bootstrap_medians, 2.5)  # 2.5th percentile
upper_bound_median = np.percentile(bootstrap_medians, 97.5)  # 97.5th percentile

print(f"95% Confidence Interval for the Median: [{lower_bound_median}, {upper_bound_median}]")
How this works:
Bootstrap Sampling: We randomly resample from the original data with replacement and calculate the mean (or median) for each resampled dataset.
Confidence Interval: We use the 2.5th and 97.5th percentiles of the bootstrap means/medians to form a 95% confidence interval.
How to change the parameter:
Population Mean: The code calculates the mean of each bootstrap sample with np.mean().
Population Median: Replace np.mean() with np.median() in the loop to calculate the median instead.
This method is flexible and can be adapted for other statistics, such as standard deviation or variance, just by changing the function used to compute the statistic in each bootstrap sample.
```

Population Parameter: This is the true value we want to estimate, such as the population mean (μ). It is typically unknown because we can't measure the entire population.

Sample Statistic: This is an estimate of the population parameter, such as the sample mean (x̄). We calculate it based on data from a sample, which represents a subset of the population.

Confidence intervals help bridge the gap between these two by providing a range of values (based on the sample statistic) that is likely to contain the population parameter with a certain level of confidence (e.g., 95%).

The sample statistic is used to estimate the population parameter.
The confidence interval reflects the uncertainty of this estimate by accounting for sampling variability.
Failing to distinguish between them could lead to confusion, as the sample statistic is not the exact population value but rather a snapshot that varies depending on the sample. The confidence interval quantifies how much the sample statistic may differ from the true population parameter.

a) What is the process of bootstrapping?
Bootstrapping is a cool way to get a better idea of how reliable your estimates are. Here’s how you can think of it:

Start with Your Sample: You begin with the data you collected.

Resample with Replacement: Randomly pick values from your dataset to create a new dataset of the same size. You can choose the same number more than once.

Calculate a Statistic: From this new dataset, find a statistic, like the average.

Repeat: Do this a lot—usually thousands of times—to get lots of different averages.

Look at the Results: Check out how the averages spread out. This helps you understand how much your estimates could change.


b) What is the main purpose of bootstrapping?
The main goal of bootstrapping is to figure out how trustworthy your estimates are. By creating lots of different versions of your dataset, you can see how much your statistics (like the average) might vary. This helps you make confidence intervals, which show where the true average might fall.


c) If you had a (hypothesized) guess about what the average of a population was, how could you use bootstrapping to assess whether your guess might be plausible?
If you have a guess about the average of a population and you have a sample, here’s how you can use bootstrapping:

Calculate the Sample Mean: First, find the average of your sample data.

Bootstrap Your Sample: Randomly draw values from your sample to create new datasets, calculating the average for each one.

Build a Range of Averages: After doing this many times, you’ll have a bunch of different averages.

Check the Range: Look at where these averages fall. If most of them are close to your guess, it means your guess could be reasonable. But if they’re all far from your guess, you might want to rethink it.

In short, bootstrapping helps you see how your sample could behave in different situations, giving you a better idea of whether your hypothesis about the average makes sense!

A confidence interval that overlaps zero implies that zero is within the range of probable values for the population mean, implying that the medicine may have no impact on average. This is because a zero mean indicates that there is no difference between the drug and the control group, implying that we cannot safely state that the medicine has an effect.

Even if the observed sample mean is not 0, it may not be statistically significant. The confidence interval represents the uncertainty in our estimate; if it contains 0, it implies that the observed impact might be attributable to random variation rather than a meaningful effect. As a result, we "fail to reject the null hypothesis" since the evidence is insufficient to rule out the possibility of no effect.

On the other hand, if the confidence interval does not overlap zero (i.e., it ranges wholly above or below zero), it means that the possible possibilities for the population mean do not include zero. In this situation, we may infer that the medicine has a large impact, allowing us to "reject the null hypothesis" that there is no effect on average. This would occur if our sample mean deviates sufficiently from zero and the data reveal persistent evidence of a meaningful impact, making it unlikely that the observed differences are attributable to chance.

The Bootstrapping Process as it is perfect for quantitative analysis

Resample the Data: Randomly select observations from your original data with replacement (meaning you can select the same observation more than once).
Calculate Mean Differences: For each bootstrap sample, calculate the mean difference between the final health score and the initial health score.

After the main bootstrapping code below, we wil calculate the confidence interval. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('Vac_data.csv')


# Strip whitespace from column names
data.columns = data.columns.str.strip()

    
# Calculate initial and final health score differences
initial_scores = data['InitialHealthScore']
final_scores = data['FinalHealthScore']
differences = final_scores - initial_scores

# Number of bootstrap samples
n_iterations = 10000
bootstrapped_means = []

# Generate bootstrap samples
for _ in range(n_iterations):
    bootstrapped_sample = np.random.choice(differences, size=len(differences), replace=True)
    bootstrapped_means.append(np.mean(bootstrapped_sample))

# Calculate confidence intervals
lower_bound = np.percentile(bootstrapped_means, 2.5)
upper_bound = np.percentile(bootstrapped_means, 97.5)

print(f"95% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")
```

The lower bound (0.90) means that even in the worst-case scenario, the vaccine likely improves health by at least 0.90 points, and in the best-case scenario, by as much as 5.60 points! Since it does not include the zero, the vaccine has a positive effect and that's for sure. 

We can now create a histogram of. the bootsprapped means as we saw in the video to show the distribution. 

```python
plt.hist(bootstrapped_means, bins=30, alpha=0.75, color='skyblue')
plt.axvline(np.mean(bootstrapped_means), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(lower_bound, color='green', linestyle='dashed', linewidth=2, label='Lower Bound (CI)')
plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Upper Bound (CI)')
plt.title('Distribution of Bootstrapped Means')
plt.xlabel('Mean Difference')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Findings: The calculated 95% confidence interval for the mean difference in health scores is [0.90, 5.60]. Since this interval does not include zero, we reject the null hypothesis that the vaccine has no effect. In conclusion, the analysis indicates that the AliTech vaccine does improve health outcomes based on the patient data provided. While this analysis provides insights into the effectiveness of the vaccine, limitations such as the small sample size and variability in individual health responses should be considered when interpreting the results.
